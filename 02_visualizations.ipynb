```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

matches = pd.read_csv('data/matches.csv')
deliveries = pd.read_csv('data/deliveries.csv')
```

```python
matches['winner'].value_counts().head(10)
```

```python
#which team won the most matches?
winner_counts = matches['winner'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=winner_counts.values, y = winner_counts.index)
plt.title('Top 10 most winning teams in IPL')
plt.xlabel('Number of Teams')
plt.ylabel('Teams')
plt.show()
```

```python
#Number of matches played per season
matches.groupby('season')['id'].count()
```

```python
#bar graph

matches_per_season = matches.groupby('season')['id'].count()
plt.figure(figsize=(12,6))
sns.barplot(x = matches_per_season.values , y = matches_per_season.index)
plt.title('Number of matches played per season')
plt.xlabel('Numbmer of Matches')
plt.ylabel('Season')
plt.show()
```

```python
#Toss decision - Bat vs Field
matches['toss_decision'].value_counts()
```

```python
#pie Chart
plt.figure(figsize=(6,6))
matches['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Toss Decision - Bat vs Field')
plt.show()
```

```python
matches['venue'].value_counts().head(10)
```

```python
#Venue that hosted the most number of matches
venue_counts = matches['venue'].value_counts().head(10)

plt.figure(figsize = (12,6))
sns.barplot(x = venue_counts.values , y = venue_counts.index)
plt.xlabel ('Number of matches played')
plt.ylabel('Venue')
plt.show()
```

```python
#Does winning toss = Winning match?

toss_win = matches[matches['toss_winner'] == matches['winner']].shape[0]
toss_loss = matches[matches['toss_winner'] != matches['winner']].shape[0]

plt.figure(figsize = (6,6))
plt.pie([toss_win, toss_loss], labels = ['Toss winner won', 'Toss winner lost'],autopct='%1.1f%%')
plt.title('Does winning toss = winning match?')
plt.show()
```
