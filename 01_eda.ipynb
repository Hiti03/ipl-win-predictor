```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv('data/matches.csv')
deliveries = pd.read_csv('data/deliveries.csv')
```

```python
print(matches.shape)
print(deliveries.columns)
```

```python
#1) MATCHES.csv
matches.head()
```

```python
matches.isnull().sum()
```

```python
#which team won the most the matches?
matches['winner'].value_counts().head(10)
```

```python
#how many toss winners won the match?
(matches['toss_winner'] == matches['winner']).sum()

#returns 554 matches out of 1095 matches, approx. 50.6%, basically 50-50 chances, so winning toss gives no advantage, therefore, weak feature for predicting the winner
```

```python
matches.groupby('season')['id'].count()
```

```python
deliveries.head(10)
```

```python
deliveries.isnull().sum()
```

```python
# Total runs scored by each team across all IPL matches
deliveries.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False).head(10)
```

```python
#most runs by a batsman
deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(15)
```

```python
#most wickets taken by a bowler
deliveries[deliveries['is_wicket']==1].groupby('bowler')['is_wicket'].sum().sort_values(ascending=False).head(15)
```
